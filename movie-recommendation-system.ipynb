```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# merging both the datasets on 'title'
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# Retaining only useful columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking for null values
movies.isnull().sum()
```

```python
# As null count of overview is 3, which is less, so we decided to drop
movies.dropna(inplace=True)
```

```python
# checking for duplicates
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import json
json.loads(movies.iloc[0].genres)
```

```python
def convert(text):
    json_string = text
    movie_genkey = json.loads(json_string)
    s = []
    for i in movie_genkey:
        s.append(i['name'])
    return s
```

```python
# # extracting 'name' from genres
movies['genres'] = movies['genres'].apply(convert)
```

```python
# extracting 'name' from keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert1(text):
    json_string = text
    movie_cast = json.loads(json_string)
    s = []
    counter = 0
    for i in movie_cast:
        if counter==3:
            break
        s.append(i['name'])
        counter = counter + 1;
    return s
```

```python
# extracting first three characters
movies['cast'] = movies['cast'].apply(convert1)
```

```python
def convert2(text):
    json_string = text
    movie_crew= json.loads(json_string)
    s = []
    for i in movie_crew:
        if i['job']=='Director':
            s.append(i['name'])
    return s
```

```python
# extracting 'director' name
movies['crew'] = movies['crew'].apply(convert2)
```

```python
# string to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# removing spaces
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ",'') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ",'') for i in x])
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
movies = movies[['movie_id','title','tag']]
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['tag'].apply(lambda x:" ".join(x))
```

```python
movies.head(1)
```

```python
movies['tag'][0]
```

```python
movies['tag'] = movies['tag'].apply(lambda x:x.lower())
```

```python
movies.head(1)
```

```python
import nltk
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y);
```

```python
movies['tag'] = movies['tag'].apply(stem)
```

```python
movies.head(1)
```

```python
# using bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# converts each sentence into vectors
vectors = cv.fit_transform(movies['tag']).toarray()
```

```python
vectors
```

```python
# 4806 vectors and each having 5000 dimension
```

```python
# in high dimension, euclidean distance is not a good measure, we will use cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# find each vector similarity with each other vector
similarity = cosine_similarity(vectors) # similarity = 1/distance
```

```python
similarity.shape
```

```python
similarity[0] # similarity of first movie with all other movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies[movies['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
